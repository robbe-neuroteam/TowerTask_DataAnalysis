```python
from PIL import Image

# Load both images
img1 = Image.open("Figure09left.png")
img2 = Image.open("Figure09right.png")

# Make sure they have the same height
if img1.height != img2.height:
    # Resize second image to match the height of the first
    img2 = img2.resize((int(img2.width * img1.height / img2.height), img1.height))

# Create new image with combined width
total_width = img1.width + img2.width
combined_img = Image.new("RGB", (total_width, img1.height), color=(255, 255, 255))  # white background

# Paste the two images side by side
combined_img.paste(img1, (0, 0))
combined_img.paste(img2, (img1.width, 0))

# Save the result
combined_img.save("Figure09.png")

# Save the figure as a PDF
combined_img.save("Figure09.pdf", format="PDF")
#fig.savefig("Figure09.pdf", format="pdf", bbox_inches='tight', dpi=300)
```
